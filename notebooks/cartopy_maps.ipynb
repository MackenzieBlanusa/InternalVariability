### plot global 

```python
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# make the map global rather than have it zoom in to
# the extents of any plotted data
ax.set_global()

ax.stock_img()
ax.coastlines()

ax.plot(-72.5, 41.3, 'o',color='r', markersize=5)
ax.plot(-122.4, 37.7, 'o',color='r', markersize=5)
ax.plot(-19, 65, 'o',color='r', markersize=5)
ax.plot(10.5, 51, 'o',color='r', markersize=5)
ax.plot(133.7, -25.2, 'o',color='r', markersize=5)
ax.plot(-53.1, 3.9, 'o',color='r', markersize=5)

plt.show()
```

### zoom in on regions

#### US East

```python
lonmin=-80 
lonmax=-65
latmin=35
latmax=50
regional_extent = [lonmin, lonmax, latmin, latmax]
fig = plt.figure(figsize=(6,10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
ax.set_extent(regional_extent)
ax.stock_img()
# add features: Lakes, Rivers, state borders etc...
# ax.add_feature(cartopy.feature.LAND,edgecolor='k',zorder=1)
ax.add_feature(cartopy.feature.RIVERS,zorder=2)
ax.add_feature(cartopy.feature.LAKES,zorder=2)
ax.add_feature(cartopy.feature.BORDERS,zorder=2)
states = cfeature.NaturalEarthFeature(category='cultural',
                                     name='admin_1_states_provinces_lines',
                                     scale='50m',
                                     facecolor='none')
ax.add_feature(states,edgecolor='gray',zorder=3)
ax.plot(-72.5, 41.3, 'o',color='r', markersize=5)
```
